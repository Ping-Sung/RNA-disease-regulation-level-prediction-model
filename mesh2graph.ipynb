```python
def num2name(df, num):
  return str(df[(df['all_discriptor_num']  == num)]['discriptor_name'].item())
disease = 'Leukemia, Lymphocytic, Chronic, B-Cell'
traget = df[(df['root_disease']  == disease)]
num2name(traget, 'C23.550.291.500')
```

```python
def re_max_num(df, disease):
  target = list(df[(df['root_disease']  == disease)]['all_discriptor_num'])
  out = ""
  for item in target:
    if len(out) < len(item): out = item
  return str(df[(df['all_discriptor_num']  == out)]['discriptor_name'].item())
```

```python
def one_hot(index, max = 9):
  out = np.zeros(max)
  out[index-1] = 1
  return out
one_hot(3)
```

```python
def mesh2graph(disease):
  node_list = []
  edge_list = []
  edge_attr_list = []
  traget = df[(df['root_disease']  == disease)]
  root_name = re_max_num(traget, disease)
  roots = list(traget[(traget['discriptor_name']  == root_name)]['all_discriptor_num'])
  root_name = root_name.replace(',', '')
  root_name = root_name.replace('-', ' ')
  node_list.append(root_name)
  pointer_from = 0



  for root in roots:
    num_list = root.split('.')
    pointer_to = 0
    distence_from_root = 1
    for i in range(len(num_list)-1):
      pointer_from += 1
      i += 1
      node = num2name(traget, '.'.join(num_list[:-i]))
      node = node.replace(',', '')
      node = node.replace('-', ' ')
      node_list.append(node)
      edge = [pointer_from, pointer_to]
      edge_list.append(edge)
      pointer_to = pointer_from
      edge_attr_list.append(one_hot(distence_from_root))
      distence_from_root += 1
  return node_list, edge_list, edge_attr_list

node_list, edge_list, edge_attr_list = mesh2graph('Leukemia, Lymphocytic, Chronic, B-Cell')
node = np.array(node_list)
edge = np.array(edge_list)
edge_attr = np.array(edge_attr_list)
print(node)
print(edge.T)
print(edge_attr)
```

```python
disease_list = []
for i in range(len(df)):
  disease = df.iloc[i]['root_disease']
  # disease = disease.replace(',', '')
  if disease not in disease_list: disease_list.append(disease)
```

```python
node_dict = {}
edge_dict = {}
edge_attr_dict = {}
for disease in disease_list:
  node_list, edge_list, edge_attr_list = mesh2graph(disease)
  node_dict[disease] = node_list
  edge_dict[disease] = edge_list
  edge_attr_dict[disease] = edge_attr_list
seed = 0
print(node_dict[disease_list[seed]])
print(edge_dict[disease_list[seed]])
print(edge_attr_dict[disease_list[seed]])
```
